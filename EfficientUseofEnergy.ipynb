```python
import requests
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter, DayLocator

file_path = 'https://raw.github.com/abbasmoosajee07/TestDataChemEng/main/EffEnergy_Data.txt'
df = pd.read_csv(file_path, delimiter = "\t")
```

```python
# Gas and Electricity consumption on the same timeseries plot
plt.figure(figsize=(12, 6))
plt.scatter(df['datetime'], df['Electricity(kWh)'], label='Electricity Consumption', color='blue', marker='o')
plt.scatter(df['datetime'], df['Gas(kWh)'], label='Gas Consumption', color='orange', marker='o')
plt.title('Gas and Electricity Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Consumption (kWh)')
plt.legend()
plt.show()

# Gas and Electricity consumption against Temperature plot
plt.figure(figsize=(12, 6))
plt.scatter(df['temp'], df['Electricity(kWh)'], label='Electricity Consumption', color='blue', marker='o')
plt.scatter(df['temp'], df['Gas(kWh)'], label='Gas Consumption', color='orange', marker='o')
plt.title('Gas and Electricity Consumption Over Time')
plt.xlabel('Temperature(C)')
plt.ylabel('Consumption (kWh)')
plt.legend()
plt.show()
```
